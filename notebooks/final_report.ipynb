**This is the Task Report**

The task was:<br>
**1.** Implement Deep Feature Selection Model (2016, Yifeng Li et al.) Model has been implemented with Tensorflow. The code is placed in *../src/dfs/DFSModel.py*<br><br>
**2.** Estimate the DFS Model parameters. It was done via Gridsearch in *../scripts/gridsearch_dfs_model.py* and resulting model's configurations are fixed in *../summary/train_dfs.csv*<br><br>
**3.** Fit another classical features importance estimation methods (also by Gridsearch): *../scripts/gridsearch_selector_models.py*<br><br>
**4.** Trained models have been used for obtaining features importances: *../scripts/get_features_weights.py* and resulting importances are fixed in the json file: *../summary/features_weights.json*<br><br>
**5.** Fit classification models (again, the Gridsearch has been used) for all weights serieses, provided by weights estimators (DFS + classical ones)
**6.** Compare different features extractors in context of selected features subsets and applied classifiers to these subset<br><br><br>

Some remarks:<br>
Due to binary classification task with imballanced classes, roc_auc score has been used as a metric for Gridsearch models evaluation and also for this final methods comparison report<br>
In all evaluation procedures K-folds strategy was used (with K equal 10)

**Read the Report json File**

```python
REPORT = '../summary/final_report.json'

with open(REPORT) as f:
    report = json.load(f)
    
clf_names = list(report[list(report.keys())[0]].keys())
feat_names = list(report.keys())
```

**Make Plots**

```python
for clf_name in clf_names:
    
    plt.figure()
    
    for feat_name in feat_names:

        plt.plot(report[feat_name][clf_name]['n_features'], report[feat_name][clf_name]['score'])
        plt.title('Model: {}'.format(clf_name))
        plt.xlabel('Number of Features Used')
        plt.ylabel('ROS AUC')
        
    plt.legend(feat_names, title="Features Extractor", ncol=2)
        
```
